### doing dot product in numpy 

```python
import numpy as np

#feature vector
x = np.array([2,3,4])

#weight vector
w = np.array([50000,30000,-10000])

#bias
b = 20000

#prediction using dot product 
y = np.dot(w,x) + b

print("Predicted Price:", y )
```

```python
def dot_product(a,b):
    if len(a) != len(b):
       raise ValueError("Vectors must be same length")
    
    return np.dot(a,b)

```

```python
A = np.array([3, 1, 2, 0])
B = np.array([2, 3, 1, 1])
C = np.array([1, 0, 0, 3])

print(dot_product(A,B))  #recomended that same intrest like a 
print(dot_product(A,C)) # less intrest between a and c

#User A: `[3, 1, 2, 0]`
#- User B: `[2, 3, 1, 1]`
#- User C: `[1, 0, 0, 3]`

print
```
